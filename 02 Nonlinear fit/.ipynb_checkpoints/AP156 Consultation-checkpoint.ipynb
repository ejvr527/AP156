Importing necessary packages:

```python
import numpy as np
import scipy as scp
from scipy.stats import chi2
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

and generating a $\chi^2$ distribution using scipy stat chi2 package:

```python
fig, ax = plt.subplots(1, 1)
df =7
x = np.linspace(0,20,1000)
ax.plot(x, chi2.pdf(x, df),label='chi2 pdf')
#plt.legend()
plt.ylabel("density")
plt.xlabel("$\chi^2$")
plt.xticks(range(0,25,5))
plt.xlim(-0.1,20)
plt.ylim(-0.001,0.13)
plt.show()
```

To obtain the confidence interval for the value of $\chi^2_o$ from 5% to 95%, we use the percentile instead of the quantile function used in the book.

```python
alpha = [0.05,0.95]
z = chi2.pdf(x,df)

print("IOC for alpha 5% to 95%:",np.percentile(z,[5,95]))
```

which is slightly different from the IOC indicated in the book. which is (2.2 to 14.1)
